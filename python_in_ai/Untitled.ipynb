```python
# Load the iris dataset
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)
```

```python
from sklearn.preprocessing import StandardScaler



# Initialize a data standardization processor
ss = StandardScaler()

# Standardize the feature data
X = ss.fit_transform(X)
```

```python
from sklearn.decomposition import PCA


# Initialize a Principal Component Analysis (PCA) transformer, set reduced dimension to 2
pca = PCA(n_components=2)


# Apply PCA to reduce dimensionality of the data
X = pca.fit_transform(X)
```

```python
from matplotlib import pyplot as plt

# Define colors and markers for each class
colors = ['red', 'blue', 'green']
markers = ['o', '^', 's']

plt.figure(dpi=150)

# Visualize the reduced data
for i in range(len(X)):
    plt.scatter(X[i, 0], X[i, 1], c = colors[y[i]], marker = markers[y[i]])
    
plt.show()
```
